In exercise folder (same level as this notebook on github) there is **hiring.csv**. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,


**2 yr experience, 9 test score, 6 interview score**

**12 yr experience, 10 test score, 10 interview score**


<h3>Answer<h3>

53713.86 and 93747.79

```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
from google.colab import files
```

```python
# 1. Trigger the upload popup (You already did this successfully!)
uploaded = files.upload()
```

```python
# FIX: Set the file path directly to the string name of the uploaded file
file_path = "hiring.csv"
df = pd.read_csv(file_path)
```

```python
df

```

```python
# 2. Data Preprocessing & Cleaning
df['experience'] = df['experience'].fillna('zero')

word_dict = {
    'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6,
    'seven':7, 'eight':8, 'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0
}
df['experience'] = df['experience'].apply(lambda x: word_dict[x] if x in word_dict else x)

# Fix column name referencing to handle potential spacing issues
mean_test_score = math.floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mean_test_score)
```

```python
# 3. Train Multivariate Linear Regression Model
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

reg_linear = linear_model.LinearRegression()
reg_linear.fit(X, y)
```

```python
# 4. Predict Future Candidates
pred1 = reg_linear.predict([[2, 9, 6]])
pred2 = reg_linear.predict([[12, 10, 10]])
```

```python
print("\n--- MULTIVARIATE LINEAR REGRESSION RESULTS ---")
print(f"Candidate 1 (2yr exp, 9 test, 6 interview): ${pred1[0]:.2f}")
print(f"Candidate 2 (12yr exp, 10 test, 10 interview): ${pred2[0]:.2f}")
```
