# GrievX Logistic Regression Multiclass

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline

import joblib

from google.colab import files
```

```python
df = pd.read_csv(
    "expanded_historical_complaints_dataset.csv"
)

df.head()
```

```python
X = df["Problem_Name"]

y = df["Days_To_Complete"]
```

```python
def convert_days(days):

    if days <= 3:
        return "Fast"

    elif days <= 7:
        return "Medium"

    else:
        return "Slow"
```

```python
y = y.apply(convert_days)

y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=0
)
```

```python
model = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', LogisticRegression(
        multi_class='multinomial',
        max_iter=1000
    ))
])
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(y_pred)
```

```python
score = model.score(X_test, y_test)

print("Accuracy :", score)
```

```python
prediction = model.predict([
    "Poor bus service"
])

print(
    "Predicted Complaint Speed :",
    prediction[0]
)
```

```python
prediction = model.predict([
    "Hospital cleanliness complaint"
])

print(
    "Predicted Complaint Speed :",
    prediction[0]
)
```

```python
prediction = model.predict([
    "Water leakage issue"
])

print(
    "Predicted Complaint Speed :",
    prediction[0]
)
```

```python
joblib.dump(
    model,
    "grievx_logistic_multiclass_model.joblib"
)

print("Model Saved Successfully")
```
