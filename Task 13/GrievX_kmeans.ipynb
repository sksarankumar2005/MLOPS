# GrievX K-Means Clustering Project

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.cluster import KMeans

from sklearn.pipeline import Pipeline

import joblib
```

```python
df = pd.read_csv(
    "expanded_historical_complaints_dataset.csv"
)

df.head()
```

```python
X = df["Problem_Name"]
```

```python
model = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('kmeans', KMeans(
        n_clusters=5,
        random_state=0
    ))
])
```

```python
model.fit(X)
```

```python
clusters = model.predict(X)

print(clusters)
```

```python
df["Cluster"] = clusters

df.head()
```

```python
prediction = model.predict([
    "Poor bus service"
])

print(
    "Cluster :",
    prediction[0]
)
```

```python
prediction = model.predict([
    "Water leakage issue"
])

print(
    "Cluster :",
    prediction[0]
)
```

```python
prediction = model.predict([
    "Hospital cleanliness complaint"
])

print(
    "Cluster :",
    prediction[0]
)
```

```python
joblib.dump(
    model,
    "grievx_kmeans_model.joblib"
)

print("Model Saved Successfully")
```
